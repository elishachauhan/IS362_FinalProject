```python
april_plays = pd.read_csv(r'C:\Data\April.csv', index_col = False)
april_plays
```

```python
april_plays.dropna()
```

```python
april_plays.dtypes
```

```python
april_plays[['Position', 'Streams']] = april_plays[['Position','Streams']].apply(pd.to_numeric, errors = 'coerce')

april_plays.dtypes
```

```python
april_plays['Streams'].sum()
```

```python
april_plays['Streams'].mean()
```

```python
sum_of_plays = april_plays.groupby('Track Name')['Streams'].sum()
sum_of_plays
```

```python
avg_of_plays = april_plays.groupby('Track Name')['Streams'].mean()
avg_of_plays
```

```python
april_sum = sum_of_plays.sort_values(ascending=False).head(1)
april_sum
```

```python
april_avg = avg_of_plays.sort_values(ascending=False).head(1)
april_avg
```
