##### Import library modules:

```python
import pandas as pd
import numpy as np
import datetime
import json
import urllib.request
from bokeh.charts import Bar
from bokeh.plotting import output_notebook, show
from pandas.io.json import json_normalize
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

##### Create pandas DataFrame that reads Spotify data for US Charts from April 2017:

```python
april_plays = pd.read_csv(r'https://raw.githubusercontent.com/elishachauhan/IS362_FinalProject/master/April.csv', index_col = False)
april_plays
```

###### Drop values in which data is missing:

```python
april_plays.dropna()
```

##### Confirm datatypes for each column:

```python
april_plays.dtypes
```

##### Convert Position and Streams columns from object to numeric values:

```python
april_plays[['Position','Streams']] = april_plays[['Position', 'Streams']].apply(pd.to_numeric, errors ='coerce')

april_plays.dtypes
```

##### Display the index of columns:

```python
april_plays.columns
```

##### Shows statistical analysis of Streams column:

```python
april_plays['Streams'].describe()
```

##### Compute the sum of Streams column:

```python
april_plays['Streams'].sum()
```

##### Compute the mean of Streams column:

```python
april_plays['Streams'].mean()
```

##### Show the total plays for each track:

```python
sum_of_plays = april_plays.groupby('Track Name')['Streams'].sum()
sum_of_plays
```

##### Show the average plays for each track:

```python
avg_of_plays = april_plays.groupby('Track Name')['Streams'].mean()
avg_of_plays
```

##### Sort and find the top streamed track of April 2017:

```python
april_sum = sum_of_plays.sort_values(ascending=False).head(1)
april_sum
```

##### Sort and find the top track with highest averaged plays of April 2017:

```python
april_avg = avg_of_plays.sort_values(ascending=False).head(1)
april_avg
```

##### Utilize Bokeh to display visualizations for the first 3 tracks:

```python
output_notebook()
bar = Bar(april_plays.head(3), 'Track Name', values='Streams')
show(bar)
```

## Final Analysis:

After thorough analysis, music tastes can change by a month by month basis as shown in the monthly breakdown of the top tracks of 2017 so far. On a similar note, 2015 and 2016 music tastes has also changed from one year to another. Someone may ask: why does this happen? Social media and advertising has shifted the means in which music is spread and consumed. In other words, new releases are spread through electronic means at a faster rate than ever before. Recently, Spotify has made Fridays their new release days for singles and albums to accomodate their weekly chart numbers. As shown in each statistical breakdown and visualization, HUMBLE. by Kendrick Lamar was the most played song of April 2017. Since the single was released a few weeks before his new album released in April, people revisited the track even more after the album release. This pattern in marketing has changed dramatically after music streaming has grown in popular. Similar to Kendrick, some artists may want to release singles to give a taste of their album before its release. However, many artists may surprise fans and release an album or song without any indication that they were going to release anything. 

In relation to the project itself, it was tedious in cleaning up and tidying the data since there were many CSV files to load onto the pandas DataFrame. After outlining the data we wanted to retrieve, however, it was easier to find, sort, and aggregate values to determine which tracks were the most streamed for a particular month in 2017 and for the years 2015 & 2016. As for statistical analysis, utiizing the describe function and aggregating the Streams column provided a better picture on the breakdown of data. After finding the total plays and average plays, sorting the data with .head() and in descending order was easier to find the top tracks for 2017 and the top 3 tracks in 2015 & 2016. Importing the data into Tableau also assisted in visualizing US and Global Charts in a cohesive manner in one graph. Overall, Spotify has grown to be the leader in the music streaming world and gives insight how people's music tastes change over a period of time.
